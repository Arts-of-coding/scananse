```python
import pandas as pd

#data = pd.read_csv('unique_g_len.tsv', sep='\t', header=0) for all populations excluding ESCs
data = pd.read_csv('/ceph/rimlsfnwi/data/moldevbio/zhou/jarts/R/scRNA-seq/20210804/pseudobulk_4cells_TPM.tsv', sep='\t', header=0)
cellcounts = pd.read_csv('/ceph/rimlsfnwi/data/moldevbio/zhou/jarts/R/scRNA-seq/20210802/cellcounts_4cells.tsv', sep='\t', header=0)

# Load in the gene lengths and take the highest number
data2 = pd.read_csv('/ceph/rimlsfnwi/data/moldevbio/zhou/jarts/jupyter_notebooks/genelengths/idgl.tsv', sep='\t', header=0)
data3 = data2.groupby('ID')['LEN'].max().reset_index()

# Merge the dataframes to add lengths to the IDs
data4 = pd.merge(data, data3, on='ID')
```

```python
cellcounts
```

```python
# Display values not as scientific number but as normal float
pd.set_option('display.float_format', lambda x: '%.30f' % x)

# Exclude genes where no length is known
nan_value = float("NaN")

data4.replace("", nan_value, inplace=True)

data4.dropna(subset = ["LEN"], inplace=True)
```

```python
# adding values for the TPM count
data4["MIL"] = 1000000
data4["transcripts"] = len(data4)
data4["readlength"] = 75
data4
```

```python
# Selecting populations >100 cell in scATAC-seq + normalizing against cell number?
data4 = data4[['ID','LiCo','StCSC','MIL','transcripts','readlength','LEN']]
data4

# continue with cellcount normalization!
#cellcounts.loc[cellcounts['population'] == data4]
```

```python
# calculating the TPM
data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']] = data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']].mul(data['MIL'].values,axis=0)
data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']] = data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']].div(data['LEN'].values,axis=0)
data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']] = data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']].mul(data['readlength'].values,axis=0)
data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']] = data[['CjS','VEC','LNPCs','FCECs','CSB','CB','StC','CSSCs','MEC','LPCs','IC']].div(data['transcripts'].values,axis=0)
tpmdata = data
tpmdata
```

```python
# calculating the TPM (four cells)
data4[['LiCo','StCSC']] = data4[['LiCo','StCSC']].mul(data4['MIL'].values,axis=0)
data4[['LiCo','StCSC']] = data4[['LiCo','StCSC']].div(data4['LEN'].values,axis=0)
data4[['LiCo','StCSC']] = data4[['LiCo','StCSC']].mul(data4['readlength'].values,axis=0)
data4[['LiCo','StCSC']] = data4[['LiCo','StCSC']].div(data4['transcripts'].values,axis=0)
tpmdata = data4
tpmdata
```

```python
# subselecting the files to generate tpm value for each condition
for i in tpmdata.columns:
    if i != "ID" and i != "MIL" and i != "transcripts" and i != "LEN" and i != "readlength":
        print (i)
        final_df = tpmdata[['ID',i]]
        final_df.to_csv(str(i)+"tpm.tsv", sep="\t", index=False)
```

Code for quantile normalization of the counts

```python
import pandas as pd
import numpy as np
import qnorm
data = pd.read_csv('/ceph/rimlsfnwi/data/moldevbio/zhou/jarts/data/lako2021/merge_peaks/2021-27-06/tmp/newtable.tsv', sep='\t', header=0)
data
# removing low quality cellpopulations
data = data[['loc','LPCs','LNPCs','CjS','CSB','CB','StC','MEC','CSSCs']]
data
```

```python
##################################################################################################
```

```python
# for arm calculation only!
# joining the columns based on the arms and then doing the logtransform
data[['arm1']] = data[['LPCs','LNPCs','CjS','CSB','CB']].median(axis=1)
data[['arm2']] = data[['StC','CSSCs']].median(axis=1)
data
```

```python
# Do the logtransform ARMS

numeric_df = data.apply(lambda x: np.log(x+1) if np.issubdtype(x.dtype, np.number) and np.number != 0 else x)
numeric_df

# convert back to numeric (nessesary for the logtransform)
numeric_df[['arm1','arm2']] = numeric_df[['arm1','arm2']].astype(int)

# add the rownames for quantile normalization
numeric_df.index = numeric_df['loc']
numeric_df = numeric_df[['arm1','arm2']]
numeric_df
```

```python
# perform the quantile normalization
numericdf_quant = qnorm.quantile_normalize(numeric_df[['arm1','arm2']], ncpus=4)
numericdf_quant
```

```python
# saving the file
numericdf_quant.to_csv("quantilearms.tsv", sep="\t", index=True)
```
