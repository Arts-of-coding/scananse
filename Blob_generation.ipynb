TPM calculation of blob

```python
# Generating the median tpm from all populations except the population you want to compare it to
import pandas as pd
import numpy as np

lst = ['LPCs','LNPCs','CjS','CSB','CB','StC','MEC','CSSCs']

for i in lst:
    print (i)

    data = pd.read_csv('genelengths/unique_g_len.tsv', sep='\t', header=0)

    # display values not as scientific number but as normal float
    pd.set_option('display.float_format', lambda x: '%.30f' % x)

    # adding values for the TPM count
    data["MIL"] = 1000000
    data["transcripts"] = 16729
    data["readlength"] = 75
    # calculating the TPM
    data[lst] = data[lst].mul(data['MIL'].values,axis=0)
    data[lst] = data[lst].div(data['LEN'].values,axis=0)
    data[lst] = data[lst].mul(data['readlength'].values,axis=0)
    data[lst] = data[lst].div(data['transcripts'].values,axis=0)
    print(data)
    data2 = data[lst]
    del data2[i]
    print(data2)
    
    tpmdata = data2.median(axis=1)
    tpmdata.index = data['ID']
    tpmdata.index.rename('target_id', inplace=True)
    print(tpmdata)
    tpmdata.to_csv(str(i)+"blobtpm.tsv", sep="\t", index=True, header=['tpm'])
```
